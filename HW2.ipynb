Biểu đồ thể hiện giá đất trung bình theo tháng

```python
df1 = df.groupby('month')['price'].mean().reset_index(name='average_land_price')

# Determine the number of unique months and generate random colors
n = df1['month'].unique().__len__() + 1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

# Plotting the bar chart
plt.figure(figsize=(16, 10), dpi=80)
plt.bar(df1['month'], df1['average_land_price'], color=c, width=.5)

# Adding text labels on top of the bars
for i, val in enumerate(df1['average_land_price'].values):
    plt.text(i, val, f'{val:.2f}', horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight': 500, 'size': 12})

# Setting up axis labels and title
plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment='right')
plt.title("Biểu đồ thể hiện giá đất trung bình theo tháng của năm 2020", fontsize=22)
plt.ylabel('Average Land Price')
plt.show()
```

```python
df['pricePerSquare'] = df['price']/df['square']
```

```python
df_april = df[df['month'] == '04']
df_december = df[df['month'] == '12']
plt.figure(figsize=(10,5))

sns.histplot(df_april['pricePerSquare'], kde=True, color='orange', label='April', stat="density", bins=10)
sns.histplot(df_december['pricePerSquare'], kde=True, color='blue', label='December', stat="density", bins=40)

plt.title('Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)')
plt.xlabel('pricePerSquare')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Số lượng bất động sản phân theo số lượng bedrooms

```python
df_bedrooms = df.groupby('bedrooms').size().reset_index(name='count')

plt.figure(figsize=(10,5))
sns.barplot(x='bedrooms', y='count', data=df_bedrooms)
plt.title('Số lượng bất động sản phân theo số lượng bedrooms')
plt.show()
```

Tương quan giữa các trường thuộc tính

```python
# Gợi ý: sns.heatmap(cmap='RdYlGn', ...)
# Biểu đồ thể hiện correlation giữa các thuộc tính
corr_matrix = df[['price', 'square', 'bedrooms', 'bathrooms']].corr()

# Vẽ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', vmin=-1, vmax=1)
plt.title('Heatmap hệ số tương quan giữa các biến số thực')
plt.show()
```

```python
sns.pairplot(df[['price', 'square', 'bedrooms', 'bathrooms']], kind='scatter')
plt.suptitle('Biểu đồ phân tán giữa các biến số thực', y=1.02)
plt.show()
```

Mật độ giá nhà trung bình theo quận

```python
df_price_by_district = df.groupby('district')['pricePerSquare'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.displot(df_price_by_district['pricePerSquare'], kde=True)
plt.title('Mật độ giá nhà trung bình theo quận (VND/m2)')
plt.xlabel('Giá nhà trung bình (VND/m2)')
plt.ylabel('Density')
plt.show()
```
